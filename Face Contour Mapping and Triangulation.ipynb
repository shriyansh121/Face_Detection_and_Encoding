```python
# Importing Necessary Libraries
import cv2
import matplotlib.pyplot as plt
import mediapipe as mp
import numpy as np
import dlib
from scipy.spatial import Delaunay
```

# Using dlib Library

## Face Contour of image 

```python
def face_contour_img(image_path):
    #Load the image
    image = cv2.imread(image_path)

    #Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    #Initializing dlib detector and predictor
    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor("/Users/shriyansh/Documents/Data Science/projects/OPENCv/Face recognition/shape_predictor_68_face_landmarks.dat")

    #Detecting faces in the image
    faces = detector(gray)

    #Creating a new black frame to display the edge of face
    black_frame = np.zeros_like(image)
    #Iterating through each face
    for face in faces:
        #Getting coordinates of landmarks in the image of faces
        landmarks = predictor(gray,face)

        #Drawing jawline
        for i in range(1,17):
            pt1 = (landmarks.part(i-1).x, landmarks.part(i-1).y)
            pt2 = (landmarks.part(i).x, landmarks.part(i).y)
            cv2.line(black_frame, pt1,pt2, color = (255,255,255), thickness=1)

        #Drawing left eyebrow
        for i in range(18,22):
            pt1 = (landmarks.part(i-1).x, landmarks.part(i-1).y)
            pt2 = (landmarks.part(i).x, landmarks.part(i).y)
            cv2.line(black_frame, pt1,pt2, color = (255,255,255), thickness=1)

        #Drawing right eyebrow
        for i in range(23,27):
            pt1 = (landmarks.part(i-1).x, landmarks.part(i-1).y)
            pt2 = (landmarks.part(i).x, landmarks.part(i).y)
            cv2.line(black_frame, pt1,pt2, color = (255,255,255), thickness=1)

        #Drawing upper part of nose
        for i in range(28,31):
            pt1 = (landmarks.part(i-1).x, landmarks.part(i-1).y)
            pt2 = (landmarks.part(i).x, landmarks.part(i).y)
            cv2.line(black_frame, pt1,pt2, color = (255,255,255), thickness=1)

        #Drawing lower part of nose
        for i in range(32,36):
            pt1 = (landmarks.part(i-1).x, landmarks.part(i-1).y)
            pt2 = (landmarks.part(i).x, landmarks.part(i).y)
            cv2.line(black_frame, pt1,pt2, color = (255,255,255), thickness=1)

        #Drawing left eye
        for i in range(37,42):
            pt1 = (landmarks.part(i-1).x, landmarks.part(i-1).y)
            pt2 = (landmarks.part(i).x, landmarks.part(i).y)
            cv2.line(black_frame, pt1,pt2, color = (255,255,255), thickness=1)

        #Drawing right eye
        for i in range(43,48):
            pt1 = (landmarks.part(i-1).x, landmarks.part(i-1).y)
            pt2 = (landmarks.part(i).x, landmarks.part(i).y)
            cv2.line(black_frame, pt1,pt2, color = (255,255,255), thickness=1)

        #Drawing upper portion of lips
        for i in range(49,60):
            pt1 = (landmarks.part(i-1).x, landmarks.part(i-1).y)
            pt2 = (landmarks.part(i).x, landmarks.part(i).y)
            cv2.line(black_frame, pt1,pt2, color = (255,255,255), thickness=1)

        #Drawing lower portion of lips
        for i in range(61,68):
            pt1 = (landmarks.part(i-1).x, landmarks.part(i-1).y)
            pt2 = (landmarks.part(i).x, landmarks.part(i).y)
            cv2.line(black_frame, pt1,pt2, color = (255,255,255), thickness=1)

        #Plotting the original image with contour image one by one 
        fig,ax = plt.subplots(1,2,figsize=(14,10))

        ax[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        ax[0].axis('off')
        ax[0].set_title("Orignial Image")

        ax[1].imshow(black_frame)
        ax[1].axis('off')
        ax[1].set_title("Face Contour Image")

        plt.show()

face_contour_img("/Users/shriyansh/Documents/Data Science/projects/OPENCv/Face recognition/images/1662029829_harry-potter.jpg")
```

# Face Contour with Real Time

# Triangularity on image using dlib

```python
def triangularity_using_dlib(image_path):
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    #Initializing dlib detector and predictor
    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor("/Users/shriyansh/Documents/Data Science/projects/OPENCv/Face recognition/shape_predictor_68_face_landmarks.dat")

    faces = detector(gray)

    for face in faces:
        landmarks = predictor(gray,face)

        points = np.array([[landmarks.part(i).x, landmarks.part(i).y]for i in range(0,68)])
        delaunay = Delaunay(points)

        for i in delaunay.simplices:
            pt1,pt2,pt3 = points[i[0]], points[i[1]], points[i[2]]
            cv2.line(image, tuple(pt1), tuple(pt2), color=(255,255,0), thickness=1)
            cv2.line(image, tuple(pt2), tuple(pt3), color=(255,255,0), thickness=1)
            cv2.line(image, tuple(pt3), tuple(pt1), color=(255,255,0), thickness=1)

    plt.figure(figsize=(10,10))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title("Triangularity on faces using dlib")

triangularity_using_dlib("/Users/shriyansh/Documents/Data Science/projects/OPENCv/Face recognition/images/1662029829_harry-potter.jpg")
```

# Triangularity in Real Time Face using dlib

```python
cap = cv2.VideoCapture(0)

while True:
    r, frame = cap.read()

    detector = dlib.get_frontal_face_detector()
    predictor = dlib.shape_predictor("/Users/shriyansh/Documents/Data Science/projects/OPENCv/Face recognition/shape_predictor_68_face_landmarks.dat")

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = detector(gray)

    for face in faces:
        landmarks = predictor(gray, face)

        points = np.array([[landmarks.part(i).x, landmarks.part(i).y] for i in range(0, 68)])
        delaunay = Delaunay(points)

        for i in delaunay.simplices:
            pt1, pt2, pt3 = points[i[0]], points[i[1]], points[i[2]]
            cv2.line(frame, tuple(pt1), tuple(pt2), color=(255, 255, 100), thickness=1)
            cv2.line(frame, tuple(pt3), tuple(pt2), color=(255, 255, 100), thickness=1)
            cv2.line(frame, tuple(pt1), tuple(pt3), color=(255, 255, 100), thickness=1)

    cv2.imshow("Frame", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break

cap.release()
cv2.destroyAllWindows()
```

# Triangularity on image using Mediapipe

```python
mp_face_mesh = mp.solutions.face_mesh
mp_drawing = mp.solutions.drawing_utils

image = cv2.imread("/Users/shriyansh/Documents/Data Science/projects/OPENCv/Face recognition/images/1662029829_harry-potter.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

face_mesh = mp_face_mesh.FaceMesh(
    max_num_faces=4,
    refine_landmarks=True,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)

draw_connection = mp_drawing.DrawingSpec(thickness=2, circle_radius=1,color = (0,255,0))

faces = face_mesh.process(image)
if faces.multi_face_landmarks:
    for i in faces.multi_face_landmarks:
        mp_drawing.draw_landmarks(
            image = image,
            landmark_list=i,
            connections = mp_face_mesh.FACEMESH_TESSELATION,
            landmark_drawing_spec=None,
            connection_drawing_spec=draw_connection
        )

plt.imshow(image)
plt.axis('off')
plt.title("Triangularity on image using Mediapipe.")
```
